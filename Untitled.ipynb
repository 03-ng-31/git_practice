```python
import json
from typing import List, Dict

class JSONLoader:
    def __init__(self, schema: Dict):
        self.schema = schema
    
    def load(self, json_data: str) -> Dict:
        data = json.loads(json_data)
        self.validate(data)
        return data
    
    def validate(self, data: Dict):
        # A simple validation against schema can be done here.
        # For now, we assume the schema is correct and skip detailed validation.
        pass
    
    def extract_metadata(self, data: Dict) -> Dict:
        metadata = {
            "general": self.extract_general_info(data),
            "endpoints": self.extract_endpoints(data),
            "definitions": self.extract_definitions(data)
        }
        return metadata

    def extract_general_info(self, data: Dict) -> Dict:
        info = data.get("info", {})
        return {
            "title": info.get("title"),
            "description": info.get("description"),
            "version": info.get("version"),
            "host": data.get("host"),
            "basePath": data.get("basePath"),
            "schemes": data.get("schemes"),
            "consumes": data.get("consumes"),
            "produces": data.get("produces"),
            "tags": data.get("tags")
        }

    def extract_endpoints(self, data: Dict) -> List[Dict]:
        paths = data.get("paths", {})
        endpoints = []
        for path, methods in paths.items():
            for method, details in methods.items():
                endpoints.append({
                    "path": path,
                    "method": method.upper(),
                    "operationId": details.get("operationId"),
                    "summary": details.get("summary"),
                    "description": details.get("description"),
                    "parameters": details.get("parameters"),
                    "responses": details.get("responses")
                })
        return endpoints

    def extract_definitions(self, data: Dict) -> Dict:
        return data.get("definitions", {})
```
